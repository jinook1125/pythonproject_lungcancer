```python
import pandas as pd

df = pd.read_csv('Lung Cancer_Ct.csv')


df.head()
```

### [ 파이썬 ]

```python
select_id = int(input("id를 입력하시오: "))

# 해당 id의 family_history 데이터를 조회합니다.
family_history_series = df.loc[df['id'] == select_id, 'family_history']
age_series = df.loc[df['id'] == select_id, 'age']

print('[family_history: Yes = 1, No = 0]')   # 첫 줄에 1과 0을 정의하기 // 이렇게 할지 아니면 판다스의 내용을 1과 0으로 바꿀지 회의하기!!
print()

if not family_history_series.empty:                       # 판다스의 공식 같은 느낌? (if not empty와 values[0]을 사용)
    family_history = family_history_series.values[0]
    age = age_series.values[0]
    
    if family_history == 'Yes':
        print(f"id: {select_id}\nAge: {age}\nfamily_history: 1")
    elif family_history == 'No':
        print(f"id:{select_id}\nAge: {age}\nfamily_history: 0")
else:
    # 검색 결과가 없는 경우
    print("해당 id를 찾을 수 없습니다.")


# 검색 결과창에 gender,coutry등을 포함 시켜야할지 회의하기!!
# 판다스로 결과창을 출력하는 방향으로 해야하나?
```

### [ 판다스 ]
