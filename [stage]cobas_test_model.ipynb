```python
import pandas as pd

path = r"G:\내 드라이브\PYTHON_PROJECT\Lung Cancer_Ct.csv"
LC = pd.read_csv(path)
```

```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact

# cancer_stage 전처리 (Stage I → I)
LC["cancer_stage"] = (
    LC["cancer_stage"]
        .astype(str)
        .str.strip()
        .str.upper()
        .str.replace("STAGE ", "", regex=False)
)

LC["cancer_stage"] = pd.Categorical(
    LC["cancer_stage"],
    categories=["I", "II", "III", "IV"],
    ordered=True
)

# 사용자 입력용 → 내부 검색용 매핑
stage_input_map = {
    "Stage I": "I",
    "Stage II": "II",
    "Stage III": "III",
    "Stage IV": "IV"
}

def show_stage_table(cancer_stage):
    # "Stage I" → "I"
    internal_stage = stage_input_map[cancer_stage]

    sub = LC[LC['cancer_stage'] == internal_stage]

    if sub.empty:
        print(f"{cancer_stage} 단계에 해당하는 데이터가 없습니다.")
    else:
        display(sub.sort_values(by='age'))  # 필요하면 다른 기준 정렬 가능

interact(
    show_stage_table,
    cancer_stage=widgets.Dropdown(
        options=["Stage I", "Stage II", "Stage III", "Stage IV"],
        value="Stage I",
        description="cancer_stage"
    )
)
```
