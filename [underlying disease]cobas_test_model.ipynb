```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# CSV 파일 경로
path = r"G:\내 드라이브\PYTHON_PROJECT\Lung Cancer_Ct.csv"
df = pd.read_csv(path)

# 드롭다운 옵션 설정
categories = {
    "고혈압 (hypertension)": "hypertension",
    "천식 (asthma)": "asthma",
    "간경화 (cirrhosis)": "cirrhosis",
    "다른암 (other_cancer)": "other_cancer"
}

dropdown = widgets.Dropdown(
    options=list(categories.keys()),
    value="고혈압 (hypertension)",
    description="질환:",
    style={"description_width": "60px"},
    layout=widgets.Layout(width="300px")
)

# 표 출력 함수
def show_table(selected_label):
    col = categories[selected_label]
    filtered = df[df[col] == 1].sort_values(by="id", ascending=True)
    if filtered.empty:
        print(f"❌ {selected_label} 환자가 없습니다.")
    else:
        display(filtered.style.hide(axis="index").set_table_styles([
            {"selector": "th", "props": "background:#f0f0f0; font-weight:bold; text-align:center;"},
            {"selector": "td", "props": "border:1px solid #ddd; padding:5px;"},
            {"selector": "table", "props": "border-collapse:collapse; width:100%;"}
        ]))

# 드롭다운과 표 연동
widgets.interact(show_table, selected_label=dropdown)
```
